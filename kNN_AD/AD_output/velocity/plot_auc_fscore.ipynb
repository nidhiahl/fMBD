```python
import pandas as pd
```

```python
auc=pd.read_csv('cumulative_auc.csv',sep=' ',skiprows=1,usecols=['euclidean', 'average', 'min', 'max'])
fscore=pd.read_csv('cumulative_fscore.csv',sep=' ',skiprows=1,usecols=['euclidean', 'average', 'min', 'max'])
```

```python
#auc
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = auc.index
y_euclidean = auc['euclidean']
y_MBD_avg = auc['average']
y_MBD_min = auc['min']
y_MBD_max = auc['max']
# Create the plot
plt.figure(figsize=(5, 3))
plt.plot(x, y_euclidean, label='euclidean_distance', color='blue', alpha=0.5)
plt.plot(x, y_MBD_avg, label='MBD_average_across_10runs', color='red', alpha=0.5)
#plt.plot(x, y_MBD_min, label='Data', color='blue', alpha=0.5)
#plt.plot(x, y_MBD_max, label='Data', color='blue', alpha=0.5)

# Fill the area between y_mean  y_std
plt.fill_between(x, y_MBD_min, y_MBD_max, color='red', alpha=0.2, label="Deviation across 10 runs")

# Add labels and title
plt.xlabel('k-value')
plt.ylabel('AUC-score')
plt.title('AUC plot for euclidean vs MBD')
plt.legend()
plt.savefig('auc_plot',format='pdf')
# Show the plot
plt.show()
```
